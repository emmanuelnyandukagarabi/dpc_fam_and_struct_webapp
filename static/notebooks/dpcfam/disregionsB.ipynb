**Goal** : Convert `disregions.txt` to `disregionsB.csv`

```python
# 0. Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Path to file
# Input file
source_path = "/u/mdmc/enyanduk/internship_areasciencepark/Data/dpcfam/dpcfam_b/zenodo_unzipped_folders/B_metaclusters_xml/mc_info_from_xml/disregions.txt"
# Output file
target_path = "/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/DPCFamB/"
```

```python
# Type of file ?
with open(source_path, "r",encoding="utf-8") as f:
  print(repr(f.readline()))
```

`Oberservation` : It's a `spaces`-separated file.

```python
# Convert to a dataframe
df = pd.read_csv(source_path, sep=r"\s+",header=None, names=["MCID","%DIS"], skip_blank_lines=True)
df["MCID"] = df["MCID"].astype(int)
df = df.sort_values(by="MCID").reset_index(drop=True)
df.head()
```

```python
# Infos
df.info()
```

```python
# Statistics
df.describe().T
```

```python
# Save to csv 
final_target = target_path + "disregionsB.csv"
df.to_csv(final_target, index=False)
```
