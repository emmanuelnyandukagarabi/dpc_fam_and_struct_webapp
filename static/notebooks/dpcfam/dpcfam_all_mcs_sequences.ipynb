```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ydata_profiling import ProfileReport
```

```python
# DPCFam Standard Mcs Sequences
df1 = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/dpcfam_sequences.csv")
df1.head()
```

```python
# DPCFamB Mcs Sequences
df2 = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/DPCFamB/dpcfamB_sequences.csv")
df2.head()
```

```python
# We concatenate the two dataframes
df = pd.concat([df1, df2], ignore_index=True).drop_duplicates().reset_index(drop=True)
df.head()
```

```python
# In mcid column,we have values like MCID where ID is an integer. We wanna sort the dataframe based on ID in ascending order
# We can extract the integer part from the mcid column and sort the dataframe based on that
df["mcid_int"] = df["mcid"].str.extract(r"MC(\d+)").astype(int)
df = df.sort_values(by="mcid_int").reset_index(drop=True)
# We can drop the mcid_int column as we don't need it anymore
df = df.drop(columns=["mcid_int"])
df.head()
```

```python
# How many unique mcids do we have?
print("Number of unique mcids:", df["mcid"].nunique())
# How many unique protein_ids do we have?
print("Number of unique protein_ids:", df["protein_id"].nunique())
# How many unique sequences do we have?
print("Number of unique sequences:", df["aa_seq"].nunique())
```

```python
# Save the dataframe to csv file
df.to_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/Merged/dpcfam_all_mcs_sequences.csv", index=False)
```
