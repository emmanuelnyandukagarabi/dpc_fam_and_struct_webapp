```python
# Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# I. Dataset Source : UniRef50

```python
df = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/uniref50_proteins.csv")
df.head()
```

```python
# We need only two columns: "uniref50_id" and "length"
df = df[["uniref50_id", "length"]]
# We rename uniref50_id to protein_id and length to protein_length
df = df.rename(columns={"uniref50_id": "protein_id", "length": "protein_length"})
df.head()
```

```python
# How many unique protein_ids do we have in uniref50_proteins.csv?
uniref50_unique_protein_ids = df["protein_id"].nunique()
print("Number of unique protein_ids in uniref50_proteins.csv:", uniref50_unique_protein_ids)
```

```python
df.info()
```

# II. Filtering UniRef50-PFam proteins in the source

```python
df_pfam = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/uniref50_pfam_valid.csv")
df_pfam.head()
```

```python
# How many unique protein_ids do we have in pfam_valid.csv?
pfam_unique_protein_ids = df_pfam["uniref50_id"].nunique()
print("Number of unique protein_ids in pfam_valid.csv:", pfam_unique_protein_ids)   
```

```python
# We  need a subdaframe of df corresponding to the unique protein_ids in pfam_valid.csv
df_pfam_unique_protein_ids = df_pfam["uniref50_id"].unique()
df_pfam_protein_lengths = df[df["protein_id"].isin(df_pfam_unique_protein_ids)]
df_pfam_protein_lengths.head()
```

```python
# Info
df_pfam_protein_lengths.info()
```

# III. Filtering DPCFam proteins in the source

```python
df_dpcfam = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCFam/Merged/dpcfam_all_mcs_sequences.csv")
df_dpcfam.head()
```

```python
# How many unique protein_ids do we have in dpcfam_all_mcs_sequences.csv?
dpcfam_unique_protein_ids = df_dpcfam["protein_id"].nunique()
print("Number of unique protein_ids in dpcfam_all_mcs_sequences.csv:", dpcfam_unique_protein_ids)   
```

```python
# We  need a subdaframe of df corresponding to the unique protein_ids in dpcfam_all_mcs_sequences.csv
df_dpcfam_unique_protein_ids = df_dpcfam["protein_id"].unique()
df_dpcfam_protein_lengths = df[df["protein_id"].isin(df_dpcfam_unique_protein_ids)]
df_dpcfam_protein_lengths.head()
```

```python
# Info
df_dpcfam_protein_lengths.info()
```

# III. DPCStruct Proteins

```python
df_dpcstruct = pd.read_csv("/u/mdmc/enyanduk/internship_areasciencepark/Dataframes/DPCStruct/dpcstruct_all_proteins_lengths.csv")
# We rename length to dpcstruct_length
df_dpcstruct = df_dpcstruct.rename(columns={"length": "protein_length"})
df_dpcstruct.head()
```

```python
# Info
df_dpcstruct.info()
```

# IV. Putting everything (Pfam, DPCFam, DPCStruct) together

```python
# Is there any intersection between dpcfam and dpcstruct protein_ids?
dpcstruct_protein_ids = set(df_dpcstruct["protein_id"])
dpcfam_protein_ids = set(df_dpcfam_protein_lengths["protein_id"])
intersection_protein_ids = dpcstruct_protein_ids.intersection(dpcfam_protein_ids)
print("Number of intersecting protein_ids between dpcfam and dpcstruct:", len(intersection_protein_ids))
# Print some of the intersecting protein_ids
print("Some intersecting protein_ids:", list(intersection_protein_ids)[:10])
```

```python
# Intersection PFam, DPCFam and DPCStruct
pfam_protein_ids = set(df_pfam_protein_lengths["protein_id"])
intersection_all = pfam_protein_ids.intersection(dpcfam_protein_ids).intersection(dpcstruct_protein_ids)
print("Number of intersecting protein_ids between pfam, dpcfam and dpcstruct:", len(intersection_all))
# Print some of the intersecting protein_ids
print("Some intersecting protein_ids between pfam, dpcfam and dpcstruct:", list(intersection_all)[:10])
```

```python
# We oncatenate the three dataframes
df = pd.concat([df_pfam_protein_lengths,df_dpcfam_protein_lengths, df_dpcstruct], ignore_index=True).drop_duplicates().reset_index(drop=True)
print(f"df shape: {df.shape}")
df.head()
```
